# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
```

# Import dataset

```python
wine = pd.read_csv('wine-composition-dataset.csv')
X = wine.iloc[:, [0, 4]].values
```

# 2

```python
kmeans = KMeans(n_clusters = 2, init = 'random', random_state = 40)
y_kmeans = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], c = 'blue', label = 'Cluster 2')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')

plt.title('Clusters of wines')
plt.xlabel('Alcohol')
plt.ylabel('Magnesium')
plt.legend()

plt.savefig('k-Mean k = 2', transparent=True)
plt.show()
```

# Elbow method

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random', random_state = 40)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, color = "m")
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.savefig('Elbow method', transparent=True)
plt.show()
```
